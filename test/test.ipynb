```python
import subprocess, tempfile, os
from pathlib import Path

url = "https://www.youtube.com/watch?v=K-IGRdxID4I&list=RDK-IGRdxID4I&start_radio=1&pp=ygULZ-y9lOuTnCBjY22gBwE%3D"  # list= 파라미터 제거 권장

tmp = tempfile.NamedTemporaryFile(suffix=".wav", delete=False)
tmp_path = Path(tmp.name)
tmp.close()

cmd = [
    "yt-dlp",
    "--no-playlist",
    "--download-sections", "*0-60",
    "-f", "bestaudio[abr<=192]/bestaudio",
    "-x",
    "--audio-format", "wav",
    "--postprocessor-args", "-ac 1 -ar 22050",
    "--force-overwrites",
    "-o", str(tmp_path),
    url,
]


print("$", " ".join(cmd))
res = subprocess.run(cmd, capture_output=True, text=True)
print("returncode:", res.returncode)
print("stdout:\n", res.stdout[:2000])
print("stderr:\n", res.stderr[:4000])

print("file:", tmp_path, "exists:", tmp_path.exists(), "size:", tmp_path.stat().st_size if tmp_path.exists() else None)
```
